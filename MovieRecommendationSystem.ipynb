```python
movies.dropna(inplace = True)
```

```python
import ast

def convert(obj):
    l = []
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies['keywords']
```

```python
movies.head(5)
```

```python
import ast

def convert3(obj):
    l = []
    c = 0
    for i in ast.literal_eval(obj):
        if c != 3:
            l.append(i['name'])
            c += 1
        else:
            break
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
import ast

def fetchDir(obj):
    l = []
    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(fetchDir)
```

```python
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ", "") for i in x])
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
    
```

```python
ps.stem('loved')
```

```python
stem(new_tags_df['tags'][0])
```

```python
new_tags_df['tags'] = new_tags_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
vectors = cv.fit_transform(new_tags_df['tags']).toarray()
```

```python
vectors
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
similarity.shape
```

```python
similarity
```

```python
sim = similarity[0]
```

```python
new_tags_df[new_tags_df['title']=="Titanic"].index[0]
```

```python
def recommend(movie):
    movie_index = new_tags_df[new_tags_df['title']==movie].index[0]
    distance = similarity[movie_index]
    movie_list = sorted(list(enumerate(distance)),reverse=True,key=lambda x :x[1])
    res = []
    for i in movie_list:
        if i[1] > 0.23:
            res.append(new_tags_df.iloc[i[0]].title)
        else:
            break
    if len(res) > 8:
        for i in range(0, 7):
            print(res[i])
    else:
        for i in range(0, len(res)):
            print(res[i])
        
            
```

```python
recommend('Batman')
```
