```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits, on = 'title')
```

```python
movies[movies['title'] == 'Me Before You']
```

```python
movies.columns
```

```python
movies = movies[['id','genres','keywords','overview','popularity','tagline','title','vote_average','vote_count','cast','crew']]
```

```python
movies.isnull().sum()
```

```python
movies['tagline'].fillna('', inplace = True)
```

```python
movies.dropna(inplace = True)
```

```python
movies.isnull().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast

def convert(obj):
    l=[]
    for i in ast.literal_eval(obj):
        l.append(i["name"])
    return l
```

```python
movies['genres']=movies['genres'].apply(convert)
```

```python
movies.iloc[0].genres
```

```python
movies['keywords']=movies['keywords'].apply(convert)
```

```python
import ast

def convert1(obj):
    l=[]
    counter =0
    for i in ast.literal_eval(obj):
        if counter < 3:
            l.append(i["name"])
            counter +=1
        else:
            break
    return l
```

```python
movies['cast']=movies['cast'].apply(convert1)
```

```python
import ast

def convert2(obj):
    l=[]

    for i in ast.literal_eval(obj):
        if i['job'] == "Director":
            l.append(i["name"])
            break
    return l
```

```python
movies['crew']=movies['crew'].apply(convert2)
```

```python
movies['overview']=movies['overview'].apply(lambda x :x.split())
```

```python
movies['tagline']=movies['tagline'].apply(lambda x :x.split())
```

```python
movies.head(1)
```

```python
movies['genres']=movies['genres'].apply(lambda x :[i.replace(" ","") for i in x])
movies['keywords']=movies['keywords'].apply(lambda x :[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x :[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x :[i.replace(" ","") for i in x])

```

```python
movies.head(1)
```

```python
movies['tags']=movies['overview'] + movies['genres'] + movies['keywords'] +movies['cast'] + movies['crew'] + movies['tagline']
```

```python
movies.head(1)
```

```python
new_df = movies[['id','title','tags']]
```

```python
new_df.head(1)
```

```python
new_df['tags']=new_df['tags'].apply(lambda x :" ".join(x))
```

```python
new_df['tags']= new_df['tags'].apply(lambda x:x.lower())
```

```python
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer('english')
def stemm(text):
    y = []
    for i in text.split():
        y.append(stemmer.stem(i))
        
    return " ".join(y)

new_df['tags'] = new_df['tags'].apply(stemm)
        
```

```python
new_df['tags']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(stop_words = 'english')

tfidf = vectorizer.fit_transform(new_df['tags']).toarray()
```

```python
tfidf.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity1 = cosine_similarity(tfidf)    
```
